```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
```

```python
arm = np.load('data/arm.npy')
```

```python
rows = 10
cols = 5
fig, ax = plt.subplots(figsize=(cols,rows))
ax.axis('off')
plt.title('title')

for i in list(range(0, min(len(arm),(rows*cols)) )):      
    a = fig.add_subplot(rows,cols,i+1)
    imgplot = plt.imshow(arm[i,:784].reshape((28,28)), cmap='gray_r', interpolation='nearest')
    plt.savefig('1.png')
    plt.xticks([])
    plt.yticks([])
```
